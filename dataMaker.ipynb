```python
for i in range(5000):
    background = AS.from_wav('Data/backgrounds/bg' + str(random.randint(1, 170)) + '.wav')
    voice = AS.from_wav('Data/voices/vc' + str(random.randint(1, 300)) + '.wav')
    background_new = background.overlay(voice, 0)
    background_new.export('Data/X/x' + str(1+i) + '.wav', format = 'wav')
    background.export('Data/Y/y' + str(1+i) + '.wav', format = 'wav')
    if (i%100) == 0:
        print(i)
```
